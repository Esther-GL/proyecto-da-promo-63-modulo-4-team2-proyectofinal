```python
import pandas as pd
import re

# 1. Aseguramos limpieza
df_aemet['description'] = df_aemet['description'].fillna('').astype(str)

# 2. REGEX "ELIMINADOR DE NULOS"
# Cambios clave:
# - Soporte para rangos (captura el segundo número si el primero falla)
# - Soporte para palabras completas: "grados", "centígrados", "kilómetros por hora"
# - Soporte para visibilidad: "100 metros" o "100 m"
regex_pro = r"""(?ix)
    (?:fuerza\s*(\d+))|                   # Caso Fuerza Beaufort
    (?:visibilidad\s*(\d+)\s*m)|          # Caso Visibilidad
    (-?\d+[.,]?\d*)\s*                    # El número (con posible negativo/decimal)
    (?:(?:\s*a\s*\d+[.,]?\d*)\s*)?        # Salto opcional de rangos (ej. '3 a 4')
    (ºC|grados|centígrados|cm|mm|km/h|l/m2|l/m²|metros|m\b) # Unidades
"""

# 3. Aplicamos
extraccion = df_aemet['description'].str.extract(regex_pro)

# 4. Consolidamos columnas (usamos la primera que encuentre dato)
df_aemet['valor_numerico'] = extraccion[0].combine_first(extraccion[1]).combine_first(extraccion[2])
df_aemet['unidad_medida'] = extraccion[3].fillna('Fuerza')

# 5. Normalización de Unidades para análisis
unit_map = {
    'ºC': 'ºC', 'grados': 'ºC', 'centígrados': 'ºC',
    'metros': 'm', 'm': 'm',
    'km/h': 'km/h', 'l/m2': 'l/m2', 'l/m²': 'l/m2'
}
df_aemet['unidad_medida'] = df_aemet['unidad_medida'].map(unit_map).fillna(df_aemet['unidad_medida'])

# 6. Conversión a float
df_aemet['valor_numerico'] = pd.to_numeric(df_aemet['valor_numerico'].str.replace(',', '.'), errors='coerce')

# --- EL CHEQUEO DEFINITIVO ---
nulos_finales = df_aemet['valor_numerico'].isna().sum()
porcentaje = (nulos_finales / len(df_aemet)) * 100
print(f"📊 Nulos restantes: {nulos_finales} ({porcentaje:.2f}%)")
```

```python
# 1. Función para extraer la provincia (quitando "Costa", "Litoral", etc.)
def limpiar_zona(nombre):
    if 'A Coruña' in nombre: return 'A Coruña'
    if 'Pontevedra' in nombre: return 'Pontevedra'
    if 'Asturias' in nombre: return 'Asturias'
    if 'Cantabria' in nombre: return 'Cantabria'
    if 'Tarragona' in nombre: return 'Tarragona'
    if 'Gipuzkoa' in nombre: return 'Gipuzkoa'
    if 'Bizkaia' in nombre: return 'Bizkaia'
    if 'Menorca' in nombre: return 'Baleares'
    # Si no coincide con las comunes, devolvemos el nombre original o lo que esté tras el guión
    if '-' in nombre:
        return nombre.split('-')[-1].strip()
    return nombre

# 2. Aplicamos la limpieza a una nueva columna
df_aemet['provincia_limpia'] = df_aemet['area_desc'].apply(limpiar_zona)

# 3. Graficamos de nuevo
plt.figure(figsize=(12, 8))
top_prov_limpias = df_aemet['provincia_limpia'].value_counts().head(15)

sns.barplot(x=top_prov_limpias.values, y=top_prov_limpias.index, 
            hue=top_prov_limpias.index, palette='viridis', legend=False)

plt.title('Ranking de Provincias Con Más Avisos', fontsize=16)
plt.xlabel('Número Total de Avisos')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Filtramos estrictamente por temperatura
df_temp = df_aemet[df_aemet['unidad_medida'] == 'ºC'].copy()

# 2. IMPORTANTE: Asegurarnos de que el valor es numérico y negativo donde deba serlo
df_temp['valor_numerico'] = pd.to_numeric(df_temp['valor_numerico'], errors='coerce')

# 3. Creamos el Pivot Table asegurando que calculamos la MEDIA (mean)
pivot_temp = df_temp.pivot_table(index='año', columns='mes_num', values='valor_numerico', aggfunc='mean')

# 4. Renombrar columnas a meses
nombres_meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
pivot_temp.columns = [nombres_meses[i-1] for i in pivot_temp.columns]

# 5. Gráfica con escala de colores de frío a calor (RdYlBu_r)
plt.figure(figsize=(14, 7))
sns.heatmap(pivot_temp, annot=True, cmap='RdYlBu_r', center=0, fmt=".1f", linewidths=.5)

plt.title('Temperatura Media de los Avisos (ºC)', fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Filtramos solo por avisos de temperatura (ºC)
# Buscamos valores positivos (calor) superiores a 35 grados
df_calor = df_aemet[(df_aemet['unidad_medida'] == 'ºC') & (df_aemet['valor_numerico'] > 30)].copy()

# 2. Obtenemos el valor máximo registrado por cada provincia
top_calor = df_calor.groupby('area_desc')['valor_numerico'].max().sort_values(ascending=False).head(15)

# 3. Visualización
plt.figure(figsize=(12, 8))
sns.barplot(x=top_calor.values, y=top_calor.index, hue=top_calor.index, palette='YlOrRd', legend=False)

plt.title('Provincias con los Avisos de Temperatura más Altos (Máximos Históricos)', fontsize=16)
plt.xlabel('Temperatura Máxima en el Aviso (ºC)')
plt.ylabel('Zona de Aviso / Provincia')
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Añadimos las etiquetas de los grados exactos al final de las barras
for i, v in enumerate(top_calor.values):
    plt.text(v + 0.5, i, f"{v}ºC", va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Filtramos por avisos de frío (ºC) 
# Usamos un umbral de < 0 para capturar heladas reales
df_frio = df_aemet[(df_aemet['unidad_medida'] == 'ºC') & (df_aemet['valor_numerico'] < 5)].copy()

# 2. Obtenemos el valor MÍNIMO registrado (el más frío) por zona
# Ordenamos de menor a mayor (ascending=True) para que los más fríos salgan arriba
top_frio = df_frio.groupby('area_desc')['valor_numerico'].min().sort_values(ascending=True).head(15)

# 3. Visualización
plt.figure(figsize=(12, 8))
# Usamos paleta 'Blues_r' para representar el frío
sns.barplot(x=top_frio.values, y=top_frio.index, hue=top_frio.index, palette='Blues_r', legend=False)

plt.title('Zonas con los Avisos de Temperatura más Bajos (Récords de Frío)', fontsize=16)
plt.xlabel('Temperatura Mínima en el Aviso (ºC)')
plt.ylabel('Zona de Aviso')
plt.grid(axis='x', linestyle='--', alpha=0.6)

# 4. Añadimos las etiquetas de los grados exactos
# Corregido: usamos top_frio en lugar de top_calor
for i, v in enumerate(top_frio.values):
    # Ajustamos la posición del texto según el valor sea negativo o positivo
    offset = 0.2 if v > 0 else -1.5
    plt.text(v + offset, i, f"{v}ºC", va='center', fontweight='bold')

plt.tight_layout()
plt.show()
```
